# Concat

```python
s1=pd.Series([1,2,3])
s2=pd.Series([4,5,6,7])
```

```python
pd.concat([s1,s2])
```

```python

df2=pd.DataFrame({'ID':[1,2,3,4],"NAME":['AAA','BBB','CCC','DDD'],'CLASS':[7,8,9,10]})
df3=pd.DataFrame({'ID':[5,6,7,8],"NAME":['EEE','FFF','GGG','HHH'],'CLASS':[1,2,3,4]})
```

```python
pd.concat([df2,df3])
```

```python
pd.concat([df3,df2],axis=1)
```

```python
pd.concat([df2,df3],axis=0,ignore_index=True)
```

```python

df4=pd.DataFrame({'ID':[1,2,3,4],"NAME":['AAA','BBB','CCC','DDD'],'CLASS':[7,8,9,10]})
df5=pd.DataFrame({'ID':[7,8],"NAME":['GGG','HHH'],'CLASS':[3,4]})
```

```python
pd.concat([df4,df5],axis=1)
```

```python
#intersection
pd.concat([df4,df5],axis=1,join="inner")
```

```python
#Labeling
pd.concat([df4,df5],keys=['df4','df5'])
```

```python
pd.concat([df4,df5],axis=1,keys=['df4','df5'])
```

```python
df6=pd.DataFrame({'ID':[1,2,3,4],"NAME":['AAA','BBB','CCC','DDD'],'CLASS':[7,8,9,10]})
df7=pd.DataFrame({'Marks':[33,34,35,36,45]})
```
